```python
key_list = [
    "Compile Query",
    "Prepare Plan",
    "Get Query Coordinator (AM)",
    "Submit Plan",
    "Start DAG",
    "Run DAG"
]

result_dict = {}

with open('beeline_consent_query_stderr.txt', 'r') as file:
    lines = file.readlines()

for line in lines:
    for key in key_list:
        if key in line:
            parts = line.split('                          ')# κενά για να γίνει ο διαχωρισμός 
            if len(parts) >= 2:
                value = parts[1].strip()
                result_dict[key] = value

print('Query Execution Summary\n',result_dict)

key_list = [
 "Map 1",
"Map 3",
"Map 4",
"Reducer 2 ",
]
times = ["DURATION(ms)", "CPU_TIME(ms)", "GC_TIME(ms)", "INPUT_RECORDS", "OUTPUT_RECORDS"]

result_dict = {}


for line in lines:
    for key in key_list:
        if key in line:
            parts = line.split()  #  διαχωρισμός σε τμήματα
            if len(parts) >= len(times) + 1:  # +1 because of the key
                value_dict = {}
                for i in range(len(times)):
                    value_dict[times[i]] = parts[i+4].strip()  # επιλογή τμημάτων 
                result_dict[key] = value_dict

print('Task Execution Summary\n',result_dict)
# Define the key_list 
key_list = [
    "org.apache.tez.common.counters.DAGCounter",
    "File System Counters",
    "org.apache.tez.common.counters.TaskCounter",
    "HIVE",
    "Shuffle Errors",
    "Shuffle Errors_Reducer_2_INPUT_Map_1",
]

# metrics dictionary
metrics = {}

# Initialize variables to track the current group and current metrics dictionary
current_group = None
current_metrics = {}

# Iterate through lines
for line in lines:  
    line = line.strip()  # Remove leading and trailing whitespace
    if any(key in line for key in key_list):
        # Check if the line matches any of the keys in the key_list
        for key in key_list:
            if key in line:
                current_group = key
                current_metrics = {}
                metrics[current_group] = current_metrics
                break
    elif current_group and ':' in line:
        # Process lines containing metrics (assuming they have a 'key: value' format but not in the case of INFO:SOMETHING) 
        metric_parts = line.split(":")
        metric_key = metric_parts[1].strip()
        metric_value = ':'.join(metric_parts[2:]).strip()
        current_metrics[metric_key] = metric_value

# Print the organized metrics dictionary
print('Detailed Metrics per task\n',metrics)

#Lionas vasilis
```
